```python
# changing the data type of one of the column

df['id'] = df['id'].astype(object)
df['host_id'] = df['host_id'].astype(object)
df.dtypes
```

# EDA
**Task 5: Data Analysis**

*Univarate Analysis*

```python
# identifying outliers

df1= df[df['price'] < 1500]

sns.boxplot(data=df1, x='price')
plt.show()
```

```python
# Price Distribution
plt.figure(figsize=(8,5))
sns.histplot(data=df1,x='price',bins=100)
plt.title("Price Distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
df1.dtypes
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='availability_365')
plt.title("availability_365 Distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
# average price per bed
df['price per bed'] = df['price']/ df['beds']

df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

# Bivarate Analysis

 **One variable dependency on another variable**

```python
sns.barplot(data=df1,x='neighbourhood_group',y='price', hue='room_type')
plt.show()
```

```python
sns.scatterplot(data=df, x='neighbourhood_group', y='price',hue ='price per bed')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.title("Locality and Review Dependency")
sns.scatterplot(data=df1 ,x='number_of_reviews_ltm',y='price',hue='neighbourhood_group')
plt.show()
```

```python
# Geographical Distribution of AirBnB listing
plt.figure(figsize=(8,5))
plt.title('Geographical Distribution of AirBnB Listing')    
sns.scatterplot(data=df,x='longitude',y='latitude',hue='room_type')
plt.show()
```

```python
# Heatmap: Correlation of one variable with other numerical column

corr= df[['latitude','longitude','price','minimum_nights','number_of_reviews','reviews_per_month','availability_365','beds']].corr()
corr
plt.figure(figsize=(8,6))
sns.heatmap(data=corr,annot=True)
plt.show()
```
